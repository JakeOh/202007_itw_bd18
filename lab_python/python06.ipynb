**조건문(Conditional Statement)**


```
if 조건식:
    조건식이 참(True)일 때 실행할 블록(문장들)
else:
    조건식이 거짓(False)일 때 실행할 블록(문장들)
```

if-else 주의할 점:
*   if, else 블록의 문장들은 들여쓰기(indentation)가 되어야 함.

````
# 정수를 입력받아서 짝수/홀수 출력
number = int(input('정수 입력>>> '))

if number % 2 == 0:  # 2로 나눈 나머지가 0이면,
    print('짝수')    # 짝수 출력
else:                # 그렇지 않다면(2로 나눈 나머지가 0이 아니라면),
    print('홀수')    # 홀수 출력.
````

else가 없는 조건문


```
if 조건식:
    조건식이 참일 때 실행할 문장들
```



````
number = -10
if number > 0:
    print('양수')
print('=== end ===')
````

````
number = -10
if number > 0:
    print('양수')
    print('=== end ===')
````

**if-elif-else 구문**

```
if 조건식1:
    조건식1이 참일 때 실행할 문장들
elif 조건식2:
    조건식2이 참일 때 실행할 문장들
...
else:
    위의 모든 조건이 거짓일 떄 실행할 문장들
```



````
number = -10

if number > 0:  # number가 0보다 크면,
    print('positive')
elif number < 0:  # number가 0보다 작다면,
    print('negative')
else:
    print('zero')
````
